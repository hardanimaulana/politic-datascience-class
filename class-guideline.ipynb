```python
df = df.rename(columns=
               {'q1_from_reading_the_tweet_which_of_the_options_below_is_most_likely_to_be_true_about_the_stance_or_outlook_of_the_tweeter_towards_the_target': 'q1_stance', 
                'q1_from_reading_the_tweet_which_of_the_options_below_is_most_likely_to_be_true_about_the_stance_or_outlook_of_the_tweeter_towards_the_target:confidence': 'q1_conf'
               })
```

```python
df = df.rename(columns=
               {'q2_which_of_the_options_below_is_true_about_the_opinion_in_the_tweet': 'q2_opinion', 
                'q2_which_of_the_options_below_is_true_about_the_opinion_in_the_tweet:confidence': 'q2_conf',
                'orig_q1_from_reading_the_tweet_which_of_the_options_below_is_most_likely_to_be_true_about_the_stance_or_outlook_of_the_tweeter_towards_the_target': 'orig_q1_stance',
                'q1_from_reading_the_tweet_which_of_the_options_below_is_most_likely_to_be_true_about_the_stance_or_outlook_of_the_tweeter_towards_the_target_gold': 'orig_q1_gold',
                'orig_q2_which_of_the_options_below_is_true_about_the_opinion_in_the_tweet': 'orig_q2'
               })
```

```python
df["q1_stance"] = df["q1_stance"].replace("AGAINST: We can infer from the tweet that the tweeter is against the target (e.g., directly or indirectly by opposing or criticizing someone/something, by supporting someone/something opposed to the target, or by echoing the stance of somebody else).", "AGAINST")
```

```python
df.iloc[:3]
```

```python
df["q1_stance"] = df["q1_stance"].replace(['FOR: We can infer from the tweet that the tweeter supports the target (e.g., directly or indirectly by supporting someone/something, by opposing or criticizing someone/something opposed to the target, or by echoing the stance of somebody else).', 
                                           'NEUTRAL: We can infer from the tweet that the tweeter has a neutral stance towards the target. (The tweet being neither favorable nor against the target is not sufficient reason for choosing this option. One reason for choosing this option is that the tweeter supports the target entity to some extent, but is also against it to some extent.)',
                                           'NONE OF THE ABOVE: There is no clue in the tweet to reveal the stance of the tweeter towards the target (support/against/neutral).'],
                                          ['FOR', 'NEUTRAL', 'NONE OF THE ABOVE'])
```
